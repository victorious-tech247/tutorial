```
#1. To Display the images and provide information about their source.

import pydicom
# Load the images
image1 = pydicom.dcmread('/content/Datican/1864590858.dcm').pixel_array
image2 = pydicom.dcmread('/content/Datican/1874946579.dcm').pixel_array

# Display the images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image1, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image2, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.show()

# Provide information about the source
print("Source of Image 1 is: [Kaggle.com]")
print("Source of Image 2 is: [Kaggle.com]")
```

```
#2. Fuse the two images and display the output.

fused_image = (image1.astype(float) + image2.astype(float)) / 2

# Display the fused image
plt.figure(figsize=(8, 5))
plt.imshow(fused_image, cmap='gray')
plt.title('Fused Image')
plt.axis('off')
plt.show()
```

```
#3. Apply a Fourier transform to both images and display the outputs.
import numpy as np
from scipy.fft import fft2, ifft2, fftshift

# Fourier Transform
f_image1 = fft2(image1)
f_image2 = fft2(image2)

# Shift the zero frequency component to the center
f_image1_shifted = fftshift(f_image1)
f_image2_shifted = fftshift(f_image2)

# Display the Fourier Transformed images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(np.log(np.abs(f_image1_shifted) + 1), cmap='gray')
plt.title('Fourier Transform of Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(np.log(np.abs(f_image2_shifted) + 1), cmap='gray')
plt.title('Fourier Transform of Image 2')
plt.axis('off')

plt.show()
```

```

#4. Perform edge enhancement on both images and display the outputs
from skimage import filters

# Edge enhancement using Sobel filter
edges_image1 = filters.sobel(image1)
edges_image2 = filters.sobel(image2)

# Display the edge-enhanced images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(edges_image1, cmap='gray')
plt.title('Edge Enhanced Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges_image2, cmap='gray')
plt.title('Edge Enhanced Image 2')
plt.axis('off')

plt.show()
```

```
#5. Rotate both images by an angle of 45 degrees and display the outputs.
from skimage.transform import rotate

# Rotate the images
rotated_image1 = rotate(image1, 45)
rotated_image2 = rotate(image2, 45)

# Display the rotated images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(rotated_image1, cmap='gray')
plt.title('Rotated Image 1 (45 degrees)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image2, cmap='gray')
plt.title('Rotated Image 2 (45 degrees)')
plt.axis('off')

plt.show()
```
